```python
Th = 0.6
Tc = 0.3
a2 = 4
a1= 0.0

def partition_ferm(a,d,T):
  Z = exp(-(2*a -1 )/T) + exp((2 - d*sqrt(2)/2)/T) + exp(-(-2*a - 1 + d*sqrt(2)/2)/T)
  return Z

def entropy_ferm(a,d,T):
  S = ((-exp(-(2*a -1 )/T)) / partition_ferm(a,d,T)) * log((exp(-(2*a -1)/T)) /(partition_ferm(a,d,T))) - (exp((2 - d*sqrt(2)/2)/T) /(partition_ferm(a,d,T))) *log(exp((2 - d*sqrt(2)/2)/T) /(partition_ferm(a,d,T))) - (exp(-(-2*a - 1+ d*sqrt(2)/2)/T) /(partition_ferm(a,d,T))) * log(exp(-(-2*a - 1 + d*sqrt(2)/2)/T) /(partition_ferm(a,d,T)))
  return S

def internalenergy_ferm(a,d,T):
  U = ((exp(-(2*a -1 )/T))/ partition_ferm(a,d,T)) *(2*a -1 ) + (exp((2 - d*sqrt(2)/2)/T) /(partition_ferm(a,d,T))) *(-(2 - d*sqrt(2)/2)) + (exp(-(-2*a - 1 + d*sqrt(2)/2)/T) /(partition_ferm(a,d,T))) *(-2*a - 1+ d*sqrt(2)/2)
  return U
```
